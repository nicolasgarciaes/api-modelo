# Porcentaje de acierto total

```python
resultados = []

contador_total=0
contador_buenas=0
for i in range(len(df_test['variable'])):
    etiquetas, probabilidades = clf.predict(df_test['variable'].iloc[i], k=3)
    indice_maximo = np.argmax(probabilidades)
    valor_max=probabilidades[indice_maximo]
    contador_total=contador_total+1
    if df_test['label'].iloc[i] in etiquetas:
        contador_buenas=contador_buenas+1

resultados.append((contador_buenas/contador_total ))

# Imprimir los resultados
for aciertos in resultados:
    print(f"Porcentaje de acierto: {aciertos}")
```
